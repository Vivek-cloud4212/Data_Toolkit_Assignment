1. What is NumPy, and why is it widely used in Python?
- NumPy is a Python library for fast numerical computing. It is widely used because it supports fast arrays, vectorized operations, and efficient mathematical functions.

2. How does broadcasting work in NumPy?
- Broadcasting allows operations between arrays of different shapes by automatically expanding smaller arrays to match the larger ones.

3. What is a Pandas DataFrame?
- A DataFrame is a 2D labeled data structure in Pandas that stores data in rows and columns.

4. Explain the use of the groupby() method in Pandas?
- groupby():- groups data based on column values and allows aggregation like sum, mean, or count.

5. Why is Seaborn preferred for statistical visualizations?
- Because it creates attractive, easy statistical plots and integrates well with Pandas.

6. What are the differences between NumPy arrays and Python lists?
- NumPy arrays are faster and store homogeneous data, while lists are slower and can store mixed data types.

7. What is a heatmap, and when should it be used?
- A heatmap is a color-coded matrix used to show patterns, often used for correlation analysis.

8. What does the term “vectorized operation” mean in NumPy?
- An operation applied to entire arrays without loops, improving speed.

9. How does Matplotlib differ from Plotly?
- Matplotlib makes static plots and Plotly creates interactive plots.

10. What is the significance of hierarchical indexing in Pandas?
- It allows multiple index levels, helping to handle complex and multi-dimensional data.

11. What is the role of Seaborn's pairplot() function?
- It shows pair-wise relationships between numerical variables using scatterplots and histograms.

12. What is the purpose of the describe() function in Pandas?
- describe():- gives summary statistics like count, mean, min, max, etc.

13. Why is handling missing data important in Pandas?
- Because missing data can distort analysis and produce incorrect results.

14. What are the benefits of using Plotly for data visualization?
- Because it is interactive, web-friendly, high-quality visualizations with zoom and hover features.

15. How does NumPy handle multidimensional arrays?
- NumPy uses ndarrays that can store data in 1D, 2D, or higher dimensions efficiently.

16. What is the role of Bokeh in data visualization?
- Bokeh creates interactive, browser-based visualizations for dashboards.

17. Explain the difference between apply() and map() in Pandas?
- map():- works on Series element-wise.
- apply():- works on Series or DataFrame rows/columns.

18. What are some advanced features of NumPy?
- Broadcasting, vectorization, linear algebra operations, random number generation, Fourier transforms.

19. How does Pandas simplify time series analysis?
- It offers date indexing, resampling, shifting, frequency conversion, and easy time-based filtering.

20. What is the role of a pivot table in Pandas?
- It summarizes and reshapes data by aggregating values based on rows and columns.

21. Why is NumPy's array slicing faster than Python's list slicing?
- NumPy stores data in contiguous memory, enabling optimized low-level operations.

22. What are some common use cases for Seaborn?
- Correlation heatmaps, distribution plots, regression plots, and categorical data visualization.


```python
# 1. How do you create a 2D NumPy array and calculate the sum of each row?

import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

row_sum = arr.sum(axis=1)
print(row_sum)
```

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame?

import pandas as pd

df = pd.DataFrame({
    'Marks': [80, 75, 90, 85]})

mean_value = df['Marks'].mean()
print(mean_value)
```

```python
# 3. A Create a scatter plot using Matplotlib?

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [5, 7, 6, 8]

plt.scatter(x, y)
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.title("Scatter Plot")
plt.show()
```

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 5, 6, 7],
    'C': [7, 8, 9, 10]
})

corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
# 5. Generate a bar plot using Plotly?

import plotly.express as px

data = {
    'Fruits': ['Apple', 'Banana', 'Mango'],
    'Quantity': [10, 15, 7]
}

fig = px.bar(data, x='Fruits', y='Quantity', title='Fruit Quantity')
fig.show()
```

```python
# 6. Create a DataFrame and add a new column based on an existing column?

import pandas as pd

data = {
    'Name': ['Ajay', 'Bijay', 'Sujay'],

    'Age': [25, 30, 22]
}

df = pd.DataFrame(data)
df['Salary'] = [50000, 60000, 55000]
print(df)
```

```python
# 7. Write a program to perform element-wise multiplication of two NumPy arrays?

import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = a * b
print(result)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib?

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]
y2 = [1, 3, 5, 7]

plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot")
plt.legend()
plt.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold?

import pandas as pd

df = pd.DataFrame({
    'Age': [12, 25, 30, 18]
})

filtered = df[df['Age'] > 20]
print(filtered)
```

```python
# 10. Create a histogram using Seaborn to visualize a distribution?

import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 12, 15, 18, 20, 22, 25]

sns.histplot(data, bins=5, kde=False)
plt.title("Histogram")
plt.show()
```

```python
# 11. Perform matrix multiplication using NumPy?

import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = A @ B     # or np.dot(A, B)
print(result)
```

```python
# 13. Create a 3D scatter plot using Plotly?

import plotly.express as px

data = {
    'x': [1, 2, 3, 4],
    'y': [10, 20, 30, 40],
    'z': [5, 7, 9, 11]
}

fig = px.scatter_3d(data, x='x', y='y', z='z', title='3D Scatter Plot')
fig.show()
```
